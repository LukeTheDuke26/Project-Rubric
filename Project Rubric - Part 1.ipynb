```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

def generate_wordcloud(df, columns):
    """
    Generate a word cloud from the given DataFrame and specified columns.
    
    This function counts the occurrences of values in the specified columns of the DataFrame and then 
    generates a word cloud based on these frequencies. Missing values are dropped and lists are split.
    
    Parameters
    ----------
    df : pandas.DataFrame
        Input DataFrame containing the data to be used for the word cloud. It is assumed that the data
        in the specified columns is semi-colon (;) separated strings.
    
    columns : list of str
        List of column names in the DataFrame to be used for generating the word cloud.
    
    Returns
    -------
    None
        The function directly plots the word cloud using matplotlib and doesn't return anything.
        
    Note
    ----
    The function assumes that the relevant libraries (WordCloud, matplotlib, Counter from collections) have been imported.
    """
    
    # Count occurrences for the pair of columns
    counts = Counter()
    for col in columns:
        # Drop missing values, split lists, and update counts
        df[col].dropna().str.split(';').apply(counts.update)
    
    # Create a WordCloud object
    wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = None, 
                min_font_size = 10).generate_from_frequencies(counts)

    # Plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.show()

# Generate word clouds for each pair
generate_wordcloud(df, ['HaveWorkedLanguage', 'WantWorkLanguage'])

```

```python
column_pairs = [
    ("HaveWorkedLanguage", "WantWorkLanguage"),
    ("HaveWorkedFramework", "WantWorkFramework"),
    ("HaveWorkedDatabase", "WantWorkDatabase"),
    ("HaveWorkedPlatform", "WantWorkPlatform"),
]
single_column = ["IDE"]

# define a function to process and print data
def process_columns(df, col1, col2=None):
    """
    Process the given columns of the DataFrame, calculate frequency and percentage, and print the results.

    This function splits the values in the specified columns of the DataFrame (assuming semi-colon ';' separated strings),
    counts the occurrences of each unique value, calculates the percentage of each occurrence,
    and prints these results in a DataFrame format. The function processes two columns if the second column is provided.

    Parameters
    ----------
    df : pandas.DataFrame
        Input DataFrame containing the data to be processed.
    
    col1 : str
        The name of the first column in the DataFrame to be processed.
    
    col2 : str, optional
        The name of the second column in the DataFrame to be processed. If None (default), only col1 is processed.
    
    Returns
    -------
    None
        The function directly prints the results and doesn't return anything.
        
    Note
    ----
    The function assumes that the relevant libraries (pandas) have been imported.
    """
    # Split the multiple answers, count the occurrences and sort in descending order
    values1 = df[col1].str.split(';').explode().value_counts().sort_values(ascending=False)
    percentage1 = (values1 / df[col1].str.split(';').explode().count() * 100).sort_values(ascending=False)

    # Creating a dataframe for better visualization
    df1 = pd.DataFrame({'Frequency': values1, 'Percentage': percentage1})
    print(f"{col1}:")
    print(df1)
    print("\n")
    
    if col2:
        values2 = df[col2].str.split(';').explode().value_counts().sort_values(ascending=False)
        percentage2 = (values2 / df[col2].str.split(';').explode().count() * 100).sort_values(ascending=False)

        # Creating a dataframe for better visualization
        df2 = pd.DataFrame({'Frequency': values2, 'Percentage': percentage2})
        print(f"{col2}:")
        print(df2)
        print("\n")

# loop over the pairs
for pair in column_pairs:
    process_columns(df, pair[0], pair[1])
    
# loop over the single column
for col in single_column:
    process_columns(df, col)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import seaborn as sns
%matplotlib inline

df = pd.read_csv('/Users/luca/Documents/Udacity - Introduction to Data Science Course Material/survey_results_public.csv')

#Only use quant variables and drop any rows with missing values
num_vars = df[['JobSatisfaction', 'CareerSatisfaction', 'HoursPerWeek', 'Salary']]

#Drop the rows with missing Job Satisfaction answers
drop_jobsatisfaction_df = num_vars.dropna(subset=['JobSatisfaction'], axis=0)

#Mean function
fill_mean = lambda col: col.fillna(col.mean())

#Fill the mean
fill_df = drop_jobsatisfaction_df.apply(fill_mean, axis=0)

#Split into explanatory and response variable
X = fill_df[['CareerSatisfaction', 'HoursPerWeek', 'Salary']]
y = fill_df['JobSatisfaction']

#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state = 42)

#instantiate
lm_model = make_pipeline(StandardScaler(), LinearRegression()) 

#Fit
lm_model.fit(X_train, y_train) 

#Predict and score the model
y_test_preds = lm_model.predict(X_test)
"The r-squared score for the model using only quantitative variables was {} on {} values.".format(r2_score(y_test, y_test_preds), len(y_test))

```
